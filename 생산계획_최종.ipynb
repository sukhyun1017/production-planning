MPS

```python
import pandas as pd
import numpy as np
#MPS of A
month=np.zeros([13,5])
def initial(x,y,z):  #x= 재고 , y=forecast , z= orders
  #재고
  month[0][2]=x
  #forecast
  for i,f in enumerate(y):
    month[i+1][0]=f
  #orders
  for i,o in enumerate(z):
    month[i+1][1]=o

def MPS(x,y,mode):  # x=lot size, y= safety stock
  #MPS,available
  for i in range(12):
    month[i+1][2]= month[i][2]-max(month[i+1][0],month[i+1][1])+month[i+1][3]
    if month[i+1][2]<y:
       if mode=="Chasing":
          month[i+1][3]=-month[i+1][2]+y
          month[i+1][2]= month[i][2]-max(month[i+1][0],month[i+1][1])+month[i+1][3]
       if mode=="Batch":
          month[i+1][3]=x
          month[i+1][2]= month[i][2]-max(month[i+1][0],month[i+1][1])+month[i+1][3]
    if  month[i+1][2]<y:
      print('lot size is too small')
  #ATP
  N=[]
  if month[1][3] > 0:  
    month[1][4]=month[1][3]+month[0][2]-month[1][1]
  for i in range(2,13):
    if month[i][3]>0:
      N.append(i)
  for k in range(len(N)-1):
      orders = 0
      for i in range (N[k+1]-N[k]):
        orders = orders + month[N[k]+i][1]
      month[N[k]][4]=month[N[k]][3]-orders
  if month[N[-1]][3] > 0:  
      orders=0
      for i in range(N[-1],13):
        orders = orders + month[i][1]
      month[N[-1]][4]=month[N[-1]][3]-orders
  for i in range(13):
      month[i][4]= max(0,month[i][4])

########################################################################################################################
x=160
y=[100,100,100,120,150,150,150,200,200,200,200,200]
z=[120,80,50,30,20,10,0,0,0,0,0,0]
initial(x,y,z)

MPS(0,0,mode="Chasing") 
df = pd.DataFrame({'': ['forecasts','orders','available','MPS',"ATP"], 'currnet':month[0] ,
                   1:month[1],2:month[2],3:month[3],4:month[4],5:month[5],6:month[6],7:month[7],
                   8:month[8],9:month[9],10:month[10],11:month[11],12:month[12]})
df
```

```python
monthA=np.zeros([13,4])

def initial_mrp_A(x,y): # x:NOW PIB ,  y: 만드는데 필요한 개수
  #재고
  monthA[0][2]=x
  #gross requirement
  for i in range(13):
    monthA[i][0]=y*month[i][3]

def MRP_A(x,y,z,mode): # x=lot size, y= safety stock , z=leadtime
    for i in range(12):
      monthA[i+1][2]= monthA[i][2]-monthA[i+1][0]+monthA[i+1][1]+monthA[i+1-z][3]
      if monthA[i+1][2]<y:
         if i+1-z<1:
             print("scheduled receipts are needed")
         else:
             if mode == 'Batch':
                monthA[i+1-z][3]= x
                monthA[i+1][2]= monthA[i][2]-monthA[i+1][0]+monthA[i+1][1]+monthA[i+1-z][3]
                if monthA[i+1][2]<y:
                   print('lot size is small')

             if mode == 'L4L' :
                monthA[i+1-z][3]= -monthA[i+1][2]+y
                monthA[i+1][2]= monthA[i][2]-monthA[i+1][0]+monthA[i+1][1]+monthA[i+1-z][3]
      
####################################################################################################################             
initial_mrp_A(160,1)
MRP_A(0,0,2,mode='L4L')

dfA = pd.DataFrame({'': ['Gross requirements','Scheduled receipts','Projected inventory balance','Planned order releases'], 
                    'currnet':monthA[0] ,1:monthA[1],2:monthA[2],3:monthA[3],4:monthA[4],5:monthA[5],6:monthA[6],7:monthA[7],
                    8:monthA[8],9:monthA[9],10:monthA[10],11:monthA[11],12:monthA[12]})
dfA
```

```python
def MPS_update(x,y,first_sale,first_manufacture,second_sale,second_manufacture,mode): # x= lot size, y= safety stock
  month[1][1]= first_sale
  month[1][3]= first_manufacture
  month[2][1]= second_sale
  month[2][3]= second_manufacture
  month[1][2]= month[0][2]-month[1][1]+month[1][3]
  month[2][2]= month[1][2]-month[2][1]+month[2][3]
  #MPS,available
  for i in range(2,12):
    month[i+1][2]= month[i][2]-max(month[i+1][0],month[i+1][1])+month[i+1][3]
    if month[i+1][2]<y:
       if mode=="Chasing":
          month[i+1][3]=-month[i+1][2]+y
          month[i+1][2]= month[i][2]-max(month[i+1][0],month[i+1][1])+month[i+1][3]
       if mode=="Batch":
          month[i+1][3]=x
          month[i+1][2]= month[i][2]-max(month[i+1][0],month[i+1][1])+month[i+1][3]

    if  month[i+1][2]<y:
      print('lot size is too small')
  #ATP
  N=[]
  if month[1][3] > 0:  
    month[1][4]=month[1][3]+month[0][2]-month[1][1]
  for i in range(2,13):
    if month[i][3]>0:
      N.append(i)
  for k in range(len(N)-1):
      orders = 0
      for i in range (N[k+1]-N[k]):
        orders = orders + month[N[k]+i][1]
      month[N[k]][4]=month[N[k]][3]-orders
  if month[N[-1]][3] > 0:  
      orders=0
      for i in range(N[-1],13):
        orders = orders + month[i][1]
      month[N[-1]][4]=month[N[-1]][3]-orders
  for i in range(13):
      month[i][4]= max(0,month[i][4])  
  
```
